### 4.find out under which team and position spending in term of salary is high

```python
salary_sum=df.groupby(['Team','Salary'])['Position'].sum()
salary_sum
```

```python
#highest salary spended 
high_salary=df[df['Salary']==df['Salary'].max()]
print("highest salary spended:",high_salary)
```

### 4.Find if there any correlation between age and salary,represent it visually

```python
#Calculating the correlation
correlation=df['Age'].corr(df['Salary'])
print(correlation)
print('correlation between age and salary is possitive so the respective variable is in direct relationship')
```

```python
#creating a scatter plot
age=df['Age']
salary=df['Salary']
plt.scatter(x=age,y=salary)
plt.title('correlation between Age and Salary')
plt.show()
```
